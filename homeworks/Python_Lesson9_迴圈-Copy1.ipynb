### break

````python
data = [1, 9, -2, 3, 6]
for element in data:
    if element < 0:
        print(element)
        break
````

### continue

````python
data = [1, 9, -2, 3, 6]
for element in data:
    if element < 0:
        continue
    print(element)
````

### pass

````python
data = [1, 9, -2, 3, 6]
for element in data:
    if element < 0:
        pass
    print(element)
````

## iterable
```
定義了 __iter__ 即為 iterable
__iter__ 會回傳 iterator 物件
```

````python
# iterable
class MyIterable:
    def __init__(self, a):
        self.a = a
        
    def __iter__(self):
        for el in self.a:
            yield el
````

## iterator
```
定義了 __next__ 即為 iterator
也可透過 iter() 做轉換
```

````python
# Iterator
iter([1,2,3,4])
next

class Counter:
    def __init__(self):
        self.i = 0
        
    def __next__(self):
        self.i += 1
        return self.i
    
    def __iter__(self):
        return self
````

## Generator
```

使用 yield 不須定義 __next__
list comperhension 用小括號的話 出來的是generator
```

````python
# Generator
# 使用 yield 不須定義 __next__
# list comperhension 用小括號的話 出來的是generator
(x for x in range(10))

def count(start, step):
    while True:
        yield start
        start += step
````

## 九九乘法表

```
輸出結果為
1 x 1 = 1  2 x 1 = 2
1 x 2 = 2  2 x 2 = 4
.          .
.          .
.          .
```

````python
n = int(input("given n = "))

for i in range(1, n + 1):
    temp = []
    for j in range(1, n + 1):
        temp.append("{} x {} = {}".format(j, i, j*i).center(12, " "))
    print("".join(temp), end="\n\n\n")
        
````
